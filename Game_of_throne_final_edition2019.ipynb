```python
# Game of Thrones
"""
    DocString 
    This is a simulation game in the gaming zone where
    you have clues in   to reach out to the final destination and save your 
    friend from the 7 kingdom and the player have  200 gold coins 
    
    Round 1 : Decide which route you would like to take.
    Round 2 : Cross the Bridge and meet Lannister.
    Round 3 : Answer 3 questions and rescue your friend and sit on the throne. 
        
    You will have 2 chances to move forward and rescue your friend.
"""

import sys
from random import shuffle

def begin():
    name = input("What is your name?\n")
    while (name == ""):
        name = input("Enter your name\n")

    friend = input("What is your friend's name you would like to save?\n")
    while (friend == ""):
        friend = input("Enter your friend's name whom you would like to save?\n")


# Let the game begin
    print("Welcome to Game of Thrones", name, ". Your journey will begin shortly!")
    print("You will have to cross the bridge to rescue your friend", friend)
    print("If that is successful, then you have the access to sit on the Throne\n")
    input(">Press Enter to proceed....")
    return name, friend


def pathShip():
    print("You selected to Sail the Ship!")
    print("You will have to give 175 gold coins which includes 100 soldiers and 20 cannon balls\n")

def pathRoad():
    print("You selected to travel throught the Road!")
    print("You will have to give 150 gold coins which includes 100 soldiers and 50 elephants and 50 horses\n")

def selectPath(path):
    if path == "Road" or path =="road" or path=="ROAD":
        pathRoad()
        return False
    elif path == "Ship" or path=="ship" or path=="SHIP":
        pathShip()
        return False
    else:
        print("Incorrect path selected, Try again.")
        return True


def question1():
    print("Who is the current King?\n1. Bran\n2. Narendra Modi\n3. Donald Trump")
    answer = input("Your Answer=")
    if ((answer == "Bran") or (answer == "BRAN") or (answer == "1") or (answer == "bran")):
        print("Bravo!! That is the correct answer")
        return True
    else:
        print("Oops... Your answer is incorrect, you have another chance to save your friend")
        answer = input("Your Answer=")
        
        if ((answer == "Bran") or (answer == "BRAN") or (answer == "1") or (answer == "bran")) :
            print("Bravo!! That is the correct answer")
            return True
        else:
            print("You are again wrong,the answer is Bran")
            return False


def question2():
    print("Who Killed the Night Walker?\n1. Thanos\n2. Arya Stark\n3. Kavish")
    answer = input("Your Answer=")
    if ((answer == "Arya Stark") or (answer=="arya") or (answer =="arya stark") or (answer =="2") or (answer == "ARYA STARK")):
        print("Bravo!! That is the correct answer")
        return True
    else:
        print("Oops... Your answer is incorrect, you have another chance to save your friend")
        answer = input("Your Answer=")
        if((answer == "Arya Stark") or (answer == "arya") or (answer =="arya stark") or (answer =="2")or (answer == "ARYA STARK")):
            print("Bravo!! That is the correct answer")
            return True
        else:
            print("You are again wrong,the answer is Arya Stark")
            return False


def question3():
    print("Who did Sansa marry first?\n1. Obama\n2. Hitler\n3. Tyrion")
    answer = input("Your Answer=")
    if ((answer == "Tyrion") or (answer =="tyrion") or (answer=="3") or (answer == "TYRION")):
        print("Bravo!! That is the correct answer")
        return True
    else:
        print("Oops... Your answer is incorrect, you have another chance to save your friend")
        answer = input("Your Answer=")
        if  ((answer == "Tyrion") or (answer =="tyrion") or (answer=="3")or (answer == "TYRION")):
            print("Bravo!! That is the correct answer")
            return True
        else:
            print("You are again wrong,the answer is Tyrion")
            return False


def question4():
    print("How many Dragons did White Walker have?\n1. Two\n2. One\n3. Three")
    answer = input("Your Answer=")
    if ((answer == "One") or (answer =="one") or (answer == "2") or (answer == "TWO")):
        print("Bravo!! That is the correct answer")
        return True
    else:
        print("Oops... Your answer is incorrect, you have another chance to save your friend")
        answer = input("Your Answer=")
        if ((answer == "One") or (answer =="one") or (answer == "2") or (answer == "TWO")):
            print("Bravo!! That is the correct answer")
            return True
        else:
            print("You are again wrong,the answer is One")
            return False
def fail():
    print(" Better luck next time..:)")
    value = input("Do you want to rescue your friend again? Yes or No ")
    if ((value == "yes") or (value == "YES") or (value == "Yes") or (value == "y") or (value == "Y")):
        return True
    else:
        return False
    
    

def main():
    name, friend = begin()
    pathDone = True
    while pathDone:
        path = input("Which path would you like to select: Ship or Road?\n")
        pathDone = selectPath(path)

    score = 0
    myQuestions = [1,2,3,4]
    print("Here are Lannister's questions and answere all correct questions in order to rescue your friend and sit on the throne\n")
    shuffle(myQuestions)
    for i in range(3):
        print("\nNext Question")
        if (myQuestions[i] == 1):
            if(question1()):
                score += 1
            else:
                print("You failed to answer the Question, you have to start from beginning\n")
                return fail()
                break
        elif (myQuestions[i] == 2):
            if(question2()):
                score += 1
            else:
                print("You failed to answer the Question, you have to start from beginning\n")
                return fail()
                break
        elif (myQuestions[i] == 3):
            if(question3()):
                score += 1
            else:
                print("You failed to answer the Question, you have to start from beginning\n")
                return fail()
                break
        else:
            if(question4()):
                score += 1
            else:
                print("You failed to answer the Question, you have to start from beginning\n")
                return fail()
                break


    if(score == 3):
        print("Yayy, you hSave answered all question correctly")
        print("You have the key to save your friend", friend)
        print("You can sit on your throne, new king",name,"!!!!")


    print ("Completed!!!")

run = True
while run:
    run = main()

print ("Finished playing")
```
